```python
import seaborn as sns
%matplotlib inline
```

```python
#loading 'tips' dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Descriptive statistics
tips.describe()
```

```python
#Histogram - 
sns.distplot(tips['total_bill'])
#if any warnings pops up, need to install other modules called 'stats models'
#this command gives histogram, kde - a kernel density estimation 
```

```python
#drop kde line
sns.distplot(tips['total_bill'], kde=False)
#histograms tells where actually most of the bills lie, the graph shows b/w $10 and $20
#kde - it's just sum of all the normal distributions around a rug plot 
```

```python
#kernel density plot
sns.kdeplot(tips['total_bill'])
```

```python
#modify bins
sns.distplot(tips['total_bill'], bins=40)
```

```python
#Bivariate plots - joint plot
#sns.jointplot(x=x_variable,y=y_variable,dataset)
sns.jointplot('total_bill', 'tip', tips)
```

```python
#kind parameter affects what's going on inside the joint plot, by default it is scatter but can change let's try hexagon
sns.jointplot('total_bill', 'tip', tips, kind='hex')
#darker hexagons has more number of data covered. The lighter the hexagon, the less data points it has
```

```python
#kind='reg - regression line
sns.jointplot('total_bill', 'tip', tips, kind='reg')
```

```python
#kind parameter affects what's going on inside the joint plot, by default it is scatter but can change it
#kde - 2-dimensional kde
sns.jointplot('total_bill', 'tip', tips, kind='kde')

#it shows the thicker density where these points match up the most
```

```python
#pairplots plots pair-wise relationships across an entire dataframe for numerical columns and it also supports a color hue argument for categorical columns
#pairplots do jointplots for every single combination of the numerical columns in the dataset
sns.pairplot(tips)
#pairplot for all the numerical columns 
```

```python
#hue argument for categorical variable in pairplot
#palette argument allows to color this with some specific color palette
sns.pairplot(tips, hue='sex', palette='coolwarm')
#it colors the plots based off of the category
```

```python
#rugplots
sns.rugplot(tips['total_bill'])
#rugplot draws a dash mark for every point on univariate distribution, it means the thicker the stack the more datapoints 
```

```python
#what is kde line in histogram
sns.distplot(tips['total_bill'])
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
#pivot table
fp=flights.pivot_table(index='month', columns='year', values='passengers')
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='purple', linewidths=3)
```

```python
#cluster map
sns.clustermap(fp)
#it will cluster rows and columns based off of their similarity, 
#For months and years, we do not see any order because most similar months put together
#basically tries to cluster columns and rows together based on their similarity and it shows years & months in this example in clusters 
#months in the y-axis they are no longer in order,we can see sep, jun aug together it's because they are most similar months being put together
#basically puts columns or rows together that are similar to each other 
```

```python
#color mapping
sns.clustermap(fp, cmap='coolwarm')
```

```python
#standardize the scale
#normalization - 0 to 1 - standard_scale=1 - earlier it was 0 to 600 , now after normalization we could see 0 to 1 
sns.clustermap(fp, cmap='coolwarm', standard_scale=1) 
#cluster plot represents low passenger count with blue color and high passenger count with red color
```
